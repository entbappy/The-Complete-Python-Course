# Problems with working in text mode

- can't work with binary files like images
- not good for other data types like int/float/list/tuples

```python
# working with binary file
with open("files/channel.png", 'r') as f:
    print(f.read())
```

```python
# working with binary file
with open("files/channel.png", 'rb') as f:
    print(f.read())
```

```python
# working with binary file
with open("files/channel.png", 'rb') as f:
    with open("files/channel_copy.png", 'wb') as wf:
        wf.write(f.read())
```

```python
# working with other data types
with open('sample.txt','w') as f:
  f.write(5)
```

```python
# working with other data types
with open('files/sample1.txt','w') as f:
  f.write("5")
```

```python
# more complex data
d = {
    'name':'Bappy',
     'age':23,
     'gender':'male'
}

with open('sample1.txt','w') as f:
  f.write(d)
```

```python
# more complex data
d = {
    'name':'Bappy',
     'age':23,
     'gender':'male'
}

with open('files/sample1.txt','w') as f:
  f.write(str(d))
```

```python

with open('files/sample1.txt','r') as f:
  s = f.read()
  print(s)
  print(type(s))
```
